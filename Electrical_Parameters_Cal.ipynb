```python
#4

voltage_wid = wid.FloatText(placeholder='230',
                   description='Enter Voltage in volts',
                   disabled=False,value = 0,
                      style={'description_width': 'initial'})

current_wid = wid.FloatText(placeholder='1',
                   description='Enter Current in Amps',
                   disabled=False, value = 0,
                      style={'description_width': 'initial'})

res_wid = wid.FloatText(placeholder='30',
                   description='Enter Resistance in Ohms',
                   disabled=False, value = 0,
                  style={'description_width': 'initial'})


display(HTML('''<b>Please input the below parameters</b>'''))
container_layout = Layout(overflow_y='auto',
                                  display='block',
                                  border='0.5px solid black')
container_layout1 = Layout(overflow_y='auto',
                                   display='block',
                                   border='0.5px solid black')

display(widgets.VBox(children=[
    HBox([voltage_wid,current_wid]),
    HBox([res_wid])],layout=container_layout1))


display(HTML('''<b>Please select the parameter you want to calculate</b>'''))

sel = widgets.Dropdown(options=['','Current', 'Voltage','Resistance'],
                                   layout=Layout(
                                       width='30%', height='30px', left='40px'),
                                   description='',
                                   style={'description_width': 'initial'})
sel_out = widgets.Output()


def on_value_change(qa):
    sel_out.clear_output()
    with sel_out:
        if sel.value == 'Current' and current_wid.value != 0:
            display(Javascript('Jupyter.notebook.clear_output([5])'))
            display(HTML('''<b>Please select other parameter to calculate since you have already given current value !!</b>'''))
            a = 1
        else:
            display(Javascript('Jupyter.notebook.execute_cells([5])'))
        if sel.value == 'Voltage' and voltage_wid.value != 0:
            display(Javascript('Jupyter.notebook.clear_output([5])'))
            display(HTML('''<b>Please select other parameter to calculate since you have already given voltage value !!</b>'''))
            b=1
        else:
            display(Javascript('Jupyter.notebook.execute_cells([5])'))
        if sel.value == 'Resistance' and res_wid.value != 0:
            display(Javascript('Jupyter.notebook.clear_output([5])'))
            display(HTML('''<b>Please select other parameter to calculate since you have already given resistance value !!</b>'''))
            d=1
        else:
            display(Javascript('Jupyter.notebook.execute_cells([5])'))
               

sel.observe(on_value_change,names=['value'])
display(sel,sel_out)
```

```python
# 6
javascript_functions = {False: "hide()", True: "show()"}
button_descriptions = {False: "Show code", True: "Hide code"}
output1 = wid.Output()


def toggle_code(state):
    """
    Toggles the JavaScript show()/hide() function on the div.input element.
    """
    output_string = "<script>$(\"div.input\").{}</script>"
    output_args = (javascript_functions[state], )
    output = output_string.format(*output_args)
    display(HTML(output))


def button_action(value):
    """
    Calls the toggle_code function and updates the button description.
    """
    state = value.new
    toggle_code(state)
    value.owner.description = button_descriptions[state]


state = False
toggle_code(state)
l_layout = wid.Layout(width='400px', height='40px')

button_toggle = wid.ToggleButton(state,
                                 description=button_descriptions[state],
                                 button_style='success'
                                 )

placeholder = wid.Label(' ')
display(wid.HBox([placeholder, button_toggle]), output1)
button_toggle.observe(button_action, "value")
```
